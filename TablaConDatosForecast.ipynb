```python
# Función para obtener el nombre corto de un producto

def short_name(prod_name):
    return "".join(word[0] for word in prod_name.split())
```

```python
# Función para filtrar df por productos de interés y renombrar columnas

def clean_df(df, prods):  
    prod_cols_map = {}
    for prod in prods:
        acronym = short_name(prod)  
        prod_cols_map[f"sold_{prod}"] = f"sold_{acronym}"
        prod_cols_map[f"prediction_{prod}"] = f"pred_{acronym}"
    
    cols = [
        c for c in df.columns 
        if (c in prod_cols_map) or (not c.startswith("sold_") and not c.startswith("prediction_"))
    ]
    
    df_filtered = df[cols].rename(columns=prod_cols_map)
    return df_filtered
```

```python
import pandas as pd

# Ensure datetime types
weather_complete["forecast_date"] = pd.to_datetime(weather_complete["forecast_date"])
weather_complete["day"] = pd.to_datetime(weather_complete["day"])

# --- 1️⃣ Filter forecast records only ---
forecasts = weather_complete[weather_complete["data_type"] == "Forecast"].copy()

# --- 2️⃣ Identify which weeks start after each forecast_date ---
# Sunday forecasts apply to the next week's Monday–Sunday period.
# So, get forecasts issued on Sunday:
forecasts = forecasts[forecasts["forecast_date"].dt.dayofweek == 6]  # 6 = Sunday

# --- 3️⃣ Compute target week (week of the 'day' that forecast refers to) ---
forecasts["target_week"] = forecasts["day"].dt.isocalendar().week
forecasts["target_year"] = forecasts["day"].dt.isocalendar().year

# --- 4️⃣ Compute issued week (week of the forecast_date) ---
forecasts["issued_week"] = forecasts["forecast_date"].dt.isocalendar().week
forecasts["issued_year"] = forecasts["forecast_date"].dt.isocalendar().year

# --- 5️⃣ Ensure forecast corresponds to the week right after the Sunday forecast ---
# Handle week wrap-around (e.g. 2023-W52 → 2024-W01)
def is_next_week(row):
    if row["issued_year"] == row["target_year"]:
        return row["target_week"] == row["issued_week"] + 1
    # Handle the new-year transition: e.g. issued_week = 52, target_week = 1
    elif row["issued_year"] + 1 == row["target_year"] and row["issued_week"] >= 52 and row["target_week"] == 1:
        return True
    return False

forecasts = forecasts[forecasts.apply(is_next_week, axis=1)]

# --- 6️⃣ Get weekday labels (for columns) ---
forecasts["weekday_name"] = forecasts["day"].dt.day_name().str[:3].str.lower()  # mon, tue, etc.

# just include predictions done for week days Monday to Saturday
forecasts = forecasts[forecasts["weekday_name"].isin(["mon", "tue", "wed", "thu", "fri", "sat"])]

# --- 7️⃣ Pivot table: one row per station + target_week ---
vars_to_pivot = ["avg_daily_all_hours","precipitation","lowest_temp","highest_temp","avg_daily_max_min"]

weekly_forecasts = (
    forecasts.pivot_table(
        index=["weather_station_id", "cluster", "target_year", "target_week"],
        columns="weekday_name",
        values=vars_to_pivot,
        aggfunc="mean"
    )
)

# --- 8️⃣ Flatten MultiIndex columns ---
weekly_forecasts.columns = [f"{var}_{day}" for var, day in weekly_forecasts.columns]
weekly_forecasts = weekly_forecasts.reset_index()

# Optional: create a combined week identifier (e.g. 202401)
weekly_forecasts["week_id"] = (
    weekly_forecasts["target_year"].astype(str)
    + weekly_forecasts["target_week"].astype(str).str.zfill(2)
)
```

```python
# Temperature features
features_pred_sold["temp_mean_week"] = features_pred_sold[[f"avg_daily_all_hours_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]].mean(axis=1)
features_pred_sold["temp_std_week"] = features_pred_sold[[f"avg_daily_all_hours_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]].std(axis=1)
features_pred_sold["temp_max_week"] = features_pred_sold[[f"highest_temp_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]].mean(axis=1)
features_pred_sold["temp_min_week"] = features_pred_sold[[f"lowest_temp_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]].mean(axis=1)

# Extremes and ranges
features_pred_sold["temp_range_mean"] = (features_pred_sold["temp_max_week"] - features_pred_sold["temp_min_week"])
features_pred_sold["days_above_30"] = (features_pred_sold[[f"highest_temp_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]] > 30).sum(axis=1)

# Precipitation
features_pred_sold["precip_total_week"] = features_pred_sold[[f"precipitation_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]].sum(axis=1)
features_pred_sold["rain_days"] = (features_pred_sold[[f"precipitation_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]] > 0).sum(axis=1)

# Heating/Cooling Degree Days (base 18°C)
temps = features_pred_sold[[f"avg_daily_all_hours_{d}" for d in ["mon","tue","wed","thu","fri","sat"]]]
features_pred_sold["cdd_week"] = np.clip(temps - 18, 0, None).sum(axis=1)
features_pred_sold["hdd_week"] = np.clip(18 - temps, 0, None).sum(axis=1)

# Interaction
features_pred_sold["temp_precip_interaction"] = features_pred_sold["temp_mean_week"] * (1 - features_pred_sold["rain_days"]/6)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

days = ["mon", "tue", "wed", "thu", "fri", "sat"]

def compute_temp_dynamics(df, prefix="avg_daily_all_hours"):
    temps = df[[f"{prefix}_{d}" for d in days]].copy()
    
    # Standard deviation (overall variability)
    df["temp_std_week"] = temps.std(axis=1)

    # Difference between first and last day
    df["temp_first_last_diff"] = temps[f"{prefix}_sat"] - temps[f"{prefix}_mon"]

    # Day-to-day differences
    diffs = temps.diff(axis=1)
    df["temp_max_jump"] = diffs.abs().max(axis=1)
    df["temp_volatility_index"] = diffs.abs().mean(axis=1)
    df["temp_increase_days"] = (diffs > 0).sum(axis=1)
    df["temp_drop_days"] = (diffs < 0).sum(axis=1)

    # Linear trend (approximate slope)
    X = np.arange(len(days)).reshape(-1, 1)
    slopes = []
    for _, row in temps.iterrows():
        y = row.values
        model = LinearRegression().fit(X, y)
        slopes.append(model.coef_[0])
    df["temp_trend"] = slopes

    # Detect reversals or abrupt pattern changes
    df["temp_reversal"] = np.sign(temps[f"{prefix}_mon"] - temps[f"{prefix}_sat"])
    df["heat_wave_start"] = (
        (temps[f"{prefix}_mon"] < 25) &
        (temps[f"{prefix}_sat"] > 30)
    ).astype(int)

    return df

# Example usage
forecast_weekly = compute_temp_dynamics(features_pred_sold.copy(), prefix="avg_daily_all_hours")
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np
import pandas as pd


def evaluate_cluster_product_tscv(concatdf, cluster, product, weather_cols, n_splits=5):
    """
    Evaluate improvement in sales prediction using time series cross-validation.
    
    Parameters
    ----------
    concatdf : pd.DataFrame
        Full dataset containing clusters, weather features, predictions, and sales.
    cluster : int
        Cluster number to filter.
    product : str
        Product code (e.g., 'CRFNR').
    weather_cols : list of str
        List of weather feature columns.
    n_splits : int
        Number of time splits for cross-validation.
    
    Returns
    -------
    dict
        Dictionary with average MAE, R², baseline MAE, and average improvement across folds.
    """
    
    # --- Filter and prepare data ---
    df_cluster = concatdf.loc[concatdf['cluster'] == cluster].copy()

    pred_col = f'pred_{product}'
    sold_col = f'sold_{product}'

    # Check columns exist
    for col in weather_cols + [pred_col, sold_col]:
        if col not in df_cluster.columns:
            raise KeyError(f"Missing column: {col}")

    # Define X and y
    X = df_cluster[weather_cols + [pred_col]].copy()
    y = df_cluster[sold_col].copy()

    # Baseline MAE (using original model predictions)
    baseline_mae = np.mean(np.abs(y - df_cluster[pred_col]))

    # --- Time Series Cross-Validation ---
    tscv = TimeSeriesSplit(n_splits=n_splits)

    mae_scores, r2_scores, improvements = [], [], []

    for train_idx, test_idx in tscv.split(X):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        # Scale features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # Fit model
        model = RandomForestRegressor(n_estimators=300, max_depth=10, random_state=42)
        model.fit(X_train_scaled, y_train)

        # Predict
        y_pred = model.predict(X_test_scaled)

        # Compute metrics for this fold
        fold_mae = mean_absolute_error(y_test, y_pred)
        fold_r2 = r2_score(y_test, y_pred)
        fold_improvement = 100 * (baseline_mae - fold_mae) / baseline_mae if baseline_mae != 0 else np.nan

        mae_scores.append(fold_mae)
        r2_scores.append(fold_r2)
        improvements.append(fold_improvement)

    # --- Summary results ---
    results = {
        "cluster": cluster,
        "product": product,
        "baseline_MAE": round(baseline_mae, 3),
        "mean_MAE": round(np.mean(mae_scores), 3),
        "std_MAE": round(np.std(mae_scores), 3),
        "mean_R2": round(np.mean(r2_scores), 3),
        "mean_improvement_%": round(np.mean(improvements), 2),
        "n_splits": n_splits
    }

    # --- Print results summary ---
    print(f"\nCluster {cluster} | Product {product}")
    print(f"Baseline MAE: {baseline_mae:.3f}")
    print(f"Avg MAE (CV): {np.mean(mae_scores):.3f} ± {np.std(mae_scores):.3f}")
    print(f"Avg R² (CV): {np.mean(r2_scores):.3f}")
    print(f"Avg Improvement: {np.mean(improvements):.2f}%")

    return results
```
